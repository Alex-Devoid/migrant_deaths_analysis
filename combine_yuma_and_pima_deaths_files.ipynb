```python
import geopandas as gpd
import os
from glob import glob
import pandas as pd
pd.set_option('display.max_columns', 500)

#import pima country medical examiner deaths
deaths = gpd.read_file("death_files/ogis_migrant_deaths_most_recent.csv")
#project deaths with epsg:4326 crs
deaths = deaths.set_crs('epsg:4326')



deaths['Date']= pd.to_datetime(deaths['Reporting Date'])

deaths['Latitude']= deaths.Latitude.astype(float)
deaths['Longitude']= deaths.Longitude.astype(float)
deaths = gpd.GeoDataFrame(deaths, geometry=gpd.points_from_xy(deaths.Longitude,deaths.Latitude))

deaths.head(5)
```

```python
#import yuma deaths
yuma_deaths = gpd.read_file("death_files/migrant_deaths_yuma_most_recent.csv")

yuma_deaths=yuma_deaths.rename(columns={"Area": "Surface Management",'Gender':'Sex','lat':'Latitude',
                                        'lng':'Longitude',
                                        'Date':'Reporting Date',
                                       'Cause of death':'OME Determined COD'})

yuma_deaths['Date']= pd.to_datetime(yuma_deaths['Reporting Date'])

#clean lon and lat
yuma_deaths['Longitude']=pd.to_numeric(yuma_deaths['Longitude'])
yuma_deaths['Longitude'] *= -1
yuma_deaths['Latitude']=pd.to_numeric(yuma_deaths['Latitude'])

yuma_deaths = gpd.GeoDataFrame(yuma_deaths, geometry=gpd.points_from_xy(yuma_deaths.Longitude,yuma_deaths.Latitude))

yuma_deaths

```

```python
# Merge yuma and pima
frames = [deaths,yuma_deaths]

result = pd.concat(frames)
#number of deaths
print(len(result))
#merge ids into one column
def merge_id(row):
    
    if pd.isnull(row['OME number']):
        return row['ML Number']
    elif pd.isnull(row['ML Number']):
        return row['OME number']
    
result['merge_id'] = result.apply(lambda row : merge_id(row),axis=1)   
```

```python

result['Sex']=result['Sex'].str.lower()
#explore male v. female death counts
result['Sex'].value_counts(dropna=False)


```

```python

#violence
mylist3 =['GUN','GSW','stab','GSW','hanging','HOMICIDE','violence']
mylist3 = [x.lower() for x in mylist3]

result['OME Determined COD']=result['OME Determined COD'].str.lower()

result['OME Determined COD'].fillna('unknown_data', inplace = True)

deaths_force = result[result['OME Determined COD'].str.contains('|'.join(mylist3))]

print(len(deaths_force))


```

```python

#violence
mylist3 =['GUN','GSW','stab','GSW','hanging','HOMICIDE','violence']
mylist3 = [x.lower() for x in mylist3]

result['OME Determined COD']=result['OME Determined COD'].str.lower()

result['OME Determined COD'].fillna('unknown_data', inplace = True)

deaths_force = result[result['OME Determined COD'].str.contains('|'.join(mylist3))]
# number of deaths due to violence
print(len(deaths_force))


```

```python


result.to_file("death_files/ogis_migrant_deaths_091021_with_yuma.geojson",driver='GeoJSON')
result.to_csv('death_files/ogis_migrant_deaths_091021_with_yuma.csv')

result['year'] = result['Date'].dt.year
result['month'] = result['Date'].dt.month

#deaths from june 2020-06-01
result_2020_6 = result[result['Date']>= pd.Timestamp('2020-06-01')]

result_2020_6.to_csv('death_files/ogis_migrant_deaths_June2020_091021_with_yuma.csv')
```
