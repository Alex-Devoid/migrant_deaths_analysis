```python
import pandas as pd
pd.set_option('display.max_columns', 500)
import geopandas as gpd 
import rasterio
```

```python


mountains = rasterio.open('../migrant_deaths_analysis/mountains/mtn_02g')



mountains.crs


```

```python
deaths = gpd.read_file("../migrant_deaths/deaths/ogis_migrant_deaths_091021_with_yuma.geojson")

deaths=deaths[deaths['Latitude'].notnull()]
deaths['Latitude']= deaths.Latitude.astype(float)
deaths['Longitude']= deaths.Longitude.astype(float)
deaths = gpd.GeoDataFrame(deaths, geometry=gpd.points_from_xy(deaths.Longitude,deaths.Latitude))
print(len(deaths))

#match mtn_02g csr 
deaths = deaths.set_crs('epsg:4326')
# deaths = deaths.to_crs('+proj=robin +lon_0=0 +x_0=0 +y_0=0 +a=6371000 +b=6371000 +units=m no_defs')
deaths['Reporting Date']= pd.to_datetime(deaths['Reporting Date'])
deaths['Reporting_year']= pd.to_datetime(deaths['Reporting Date']).dt.year



print(deaths.total_bounds)

print(deaths.total_bounds[0])
print(deaths.total_bounds[1])


```

```python


def if_mountain(row):
    
    #print(sum(1 for _ in mountains.sample([(row['geometry'].x, row['geometry'].y)])))    
    #Generator length of 1
    for val in mountains.sample([(row['geometry'].x, row['geometry'].y)]): 
        print(val)
        if val[0] == 255:
            return 0
        else:
            return val[0]
      
    
#get mountain definition for each death point
deaths['mountains'] = deaths.apply(lambda row : if_mountain(row),axis=1)
                                      
deaths['mountains']= pd.to_numeric(deaths['mountains'])                        
deaths


```

```python


deaths.to_csv('deaths_mountains.csv')

```
